***Question 2 : Explain the difference between List and Tuple in Python.***

Answer:

Both List and Tuple are Python data structures used to store multiple items. However, they differ in terms of mutability, syntax, and performance.

List	Tuple
A List is mutable, meaning its elements can be changed after creation.	A Tuple is immutable, meaning its elements cannot be changed after creation.
Created using square brackets [ ].	Created using parentheses ( ).
Items can be added, removed, or modified.	Items cannot be added, removed, or modified.
Slightly slower because it supports modifications.	Faster than a List because it is immutable.
Used when data needs to be updated frequently.	Used when data should remain constant and secure.Example

List:

fruits = ["Apple", "Banana", "Mango"]
fruits.append("Orange")
print(fruits)

Output:

['Apple', 'Banana', 'Mango', 'Orange']

Tuple:

fruits = ("Apple", "Banana", "Mango")
print(fruits)

Output:

('Apple', 'Banana', 'Mango')
Conclusion

A List is suitable when data needs to be modified, while a Tuple is preferred when data should remain fixed because it provides better performance and data safety.

***Question 3 : What is a function in Python? Why are functions useful?***

A function in Python is a reusable block of code that performs a specific task. Functions help organize programs into smaller, manageable sections, making the code easier to read, understand, and maintain. A function is defined using the def keyword and can be called whenever its task needs to be performed.                                                  
Why are Functions Useful?

1. Code Reusability

A function can be written once and used multiple times in the program, reducing duplicate code.

2. Better Code Organization

Functions divide a large program into smaller, manageable parts, making it easier to read and maintain.

3. Easier Debugging and Maintenance

Errors can be identified and fixed within a specific function without affecting the rest of the program.

4. Improves Readability

Well-named functions make the program easier to understand for other programmers.

5. Saves Time

Reusing existing functions reduces development time and increases productivity.
Conclusion:

Functions are an essential feature of Python because they promote code reusability, modularity, readability, and easier maintenance, making programs more efficient and organized.

***Question 5 : Write a Python program to check whether a number is even or odd using conditional
statements.***

The following Python program checks whether a number entered by the user is even or odd using an if-else conditional statement.



Explanation
input() takes the number from the user.
int() converts the input into an integer.
The modulus operator (%) returns the remainder after division by 2.
If num % 2 == 0, the number is Even.
Otherwise, the number is Odd.
Conclusion

This program uses the if-else conditional statement and the modulus (%) operator to determine whether a number is even or odd.

```python
# Taking a number as input
num = int(input("Enter a number: "))

# Checking whether the number is even or odd
if num % 2 == 0:
    print(num, "is an Even number.")
else:
    print(num, "is an Odd number.")
```

***Question 8 : Write a Python program to remove duplicate values from a list using a set***

```python
# Create a list with duplicate values
numbers = [10, 20, 30, 20, 40, 10, 50]

# Remove duplicates using a set
unique_numbers = list(set(numbers))

# Print the list after removing duplicates
print("List after removing duplicates:", unique_numbers)
```

Explanation
A list named numbers contains duplicate values.
The set() function converts the list into a set, automatically removing duplicates.
The list() function converts the set back into a list.
Finally, print() displays the list containing only unique values.
Conclusion

Using a set is one of the simplest and most efficient ways to remove duplicate values from a list in Python because it automatically keeps only unique elements.

***Question 9 : Write a function that returns the square of a number**

```python
# Function to return the square of a number
def square(number):
    return number * number

# Calling the function
result = square(6)

# Display the result
print("Square of the number is:", result)
```

Explanation
The square() function takes one parameter named number.
It calculates the square by multiplying the number by itself (number * number).
The return statement sends the result back to the caller.
The function is called with the value 6, and the result (36) is printed.
Conclusion

This program demonstrates how to create and use a Python function that returns the square of a given number, promoting code reusability and modular programming

***Question 10 : Write a Python program to count how many times a number appears in a list.
Example List
[2,3,4,2,5,2]***

Explanation
A list named numbers is created.
The variable num stores the number whose occurrences are to be counted.
The count() method counts how many times the number appears in the list.
The result is displayed using the print() function.
Conclusion

This program uses Python's built-in count() method to easily determine how many times a specific number appears in a list. It is a simple and efficient way to count occurrences of an element.

```python
# Create a list
numbers = [2, 3, 4, 2, 5, 2]

# Number to count
num = 2

# Count occurrences
count = numbers.count(num)

# Display the result
print("The number", num, "appears", count, "times in the list.")
```
